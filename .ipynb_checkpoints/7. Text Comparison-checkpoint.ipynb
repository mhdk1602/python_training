# Text Comparison


Text comparison, in the context of data engineering, involves the process of finding similarities or differences between textual data. The necessity of text comparison stems from various data operations like data cleaning, integration, and transformation. It is instrumental in handling inconsistent text data, identifying duplicates, and linking records across different datasets.


### Fuzzy Matching


In data engineering, fuzzy matching is a critical tool for data cleaning and preprocessing. Data often comes from various sources and might contain inconsistencies, typos, or different naming conventions. Fuzzy matching helps in identifying and linking similar records, even if they are not exactly the same. It's particularly useful in deduplication processes, entity resolution, and data integration tasks. Understanding and implementing fuzzy matching techniques can significantly enhance the quality and reliability of data pipelines.

```python
from fuzzywuzzy import fuzz

str1 = "Data Engineering"
str2 = "Date Engineering"

similarity_ratio = fuzz.ratio(str1, str2)
print(f"The similarity ratio is: {similarity_ratio}")
```

We will simulate a scenario where we have a data frame of names and ages and we are trying to match the names in the data frame with a dictionary of reference names. We then add columns to the data frame for the matched name and the match score.

```python
import pandas as pd
from fuzzywuzzy import process

# Simulated data
data = {'Name': ['John Deo', 'Jon Doe', 'Jane Dane', 'Jake Dane'], 'Age': [25, 30, 35, 40]}
df = pd.DataFrame(data)

# Reference names
names_dict = {'John Doe': 1, 'Jane Dane': 2}

# Function to get the most similar name from the dictionary
def match_name(name, list_names, min_score=0):
    max_score = -1
    max_name = ''
    for name2 in list_names:
        score = fuzz.ratio(name, name2)
        if (score > min_score) & (score > max_score):
            max_name = name2
            max_score = score
    return (max_name, max_score)

# Applying the function to get matches and scores
df['match_tuple'] = df['Name'].apply(lambda x: match_name(x, names_dict.keys(), min_score=70))
df['match_name'] = df['match_tuple'].apply(lambda x: x[0])
df['match_score'] = df['match_tuple'].apply(lambda x: x[1])

df
```

For the next example, imagine a company has two separate customer databases due to historical reasons. Over the years, data entry errors have occurred, and now the company wants to integrate these databases into a new system. Before this, they want to identify and merge duplicate entries from these databases.
<br><br>

```python
from faker import Faker
import pandas as pd

# Instantiate faker object
fake = Faker()

# Function to create a fake dataset
def create_fake_data(num_records=1000):
    data = []
    for _ in range(num_records):
        data.append([fake.first_name(), fake.last_name(), fake.email()])
    return pd.DataFrame(data, columns=['First Name', 'Last Name', 'Email'])

# Creating two datasets
db1 = create_fake_data()
db2 = create_fake_data()

# Introducing some errors and duplicates
db2.iloc[100:200, 0] = db1.iloc[50:150, 0]
db2.iloc[200:300, 1] = db1.iloc[20:120, 1]
db2.iloc[300:400, 2] = db1.iloc[10:110, 2]
```

```python
from fuzzywuzzy import fuzz

# Function to perform fuzzy matching
def fuzzy_matching(row, db2):
    max_similarity = 0
    matched_row = None
    
    for _, db2_row in db2.iterrows():
        # Calculate the similarity score for each attribute and find their average
        name_similarity = (fuzz.ratio(row['First Name'], db2_row['First Name']) + fuzz.ratio(row['Last Name'], db2_row['Last Name'])) / 2
        email_similarity = fuzz.ratio(row['Email'], db2_row['Email'])
        
        # Calculate the overall similarity score by taking a weighted average of the name similarity (weight = 0.5) and email similarity (weight = 0.5)
        overall_similarity = (name_similarity * 0.5) + (email_similarity * 0.5)
        
        # If this is the highest similarity score so far, update max_similarity and matched_row
        if overall_similarity > max_similarity:
            max_similarity = overall_similarity
            matched_row = db2_row
    
    return matched_row, max_similarity

# List to store the results
results = []

# Perform fuzzy matching for each row in db1
for _, row in db1.iterrows():
    matched_row, max_similarity = fuzzy_matching(row, db2)
    results.append([row['First Name'], row['Last Name'], row['Email'], matched_row['First Name'], matched_row['Last Name'], matched_row['Email'], max_similarity])

# Create a DataFrame from the results
result_df = pd.DataFrame(results, columns=['First Name (DB1)', 'Last Name (DB1)', 'Email (DB1)', 'First Name (DB2)', 'Last Name (DB2)', 'Email (DB2)', 'Similarity Score'])

# Analyze the results
threshold = 80  # Similarity score threshold to consider records as duplicates
duplicates_df = result_df[result_df['Similarity Score'] >= threshold]

# Print the identified duplicates
duplicates_df
```

### Levenshtein Distance

Levenshtein distance, also known as edit distance, is a measure of similarity between two strings. It is defined as the minimum number of single-character edits (insertions, deletions, or substitutions) required to change one string into the other. In the realm of data engineering, understanding and utilizing such string similarity metrics is crucial for various tasks including data cleaning (detecting and correcting typos), record linkage (identifying duplicate records across databases), and data integration (merging data from different sources).

<br>

```python
import Levenshtein as lev

str1 = "Apple Inc."
str2 = "Appl Inc."

# Get the Levenshtein distance between the two strings
distance = lev.distance(str1, str2)
print(f"Levenshtein distance between '{str1}' and '{str2}' is: {distance}")
```

```python
import pandas as pd
import Levenshtein as lev
from faker import Faker
import pandas as pd

def generate_sample_data(num_records=100):
    fake = Faker()
    data1 = []
    data2 = []

    for _ in range(num_records):
        data1.append([fake.first_name(), fake.last_name(), fake.email()])
        data2.append([fake.first_name(), fake.last_name(), fake.email()])

    db1 = pd.DataFrame(data1, columns=['First Name', 'Last Name', 'Email'])
    db2 = pd.DataFrame(data2, columns=['First Name', 'Last Name', 'Email'])

    return db1, db2

# Generate sample data using faker
db1, db2 = generate_sample_data()

# Function to perform matching based on Levenshtein distance
def levenshtein_matching(row, db2):
    max_similarity = 0
    matched_row = None

    for _, db2_row in db2.iterrows():
        # Calculate the Levenshtein distance for each attribute and find their average
        name_distance = (lev.distance(row['First Name'], db2_row['First Name']) + lev.distance(row['Last Name'], db2_row['Last Name'])) / 2
        email_distance = lev.distance(row['Email'], db2_row['Email'])

        # Calculate the overall similarity score (1 - normalized distance)
        overall_similarity = 1 - ((name_distance * 0.5) + (email_distance * 0.5)) / max(len(row['First Name']) + len(row['Last Name']), len(db2_row['First Name']) + len(db2_row['Last Name']), len(row['Email']), len(db2_row['Email']))

        # If this is the highest similarity score so far, update max_similarity and matched_row
        if overall_similarity > max_similarity:
            max_similarity = overall_similarity
            matched_row = db2_row

    return matched_row, max_similarity

# List to store the results
results = []

# Perform Levenshtein matching for each row in db1
for _, row in db1.iterrows():
    matched_row, max_similarity = levenshtein_matching(row, db2)
    results.append([row['First Name'], row['Last Name'], row['Email'], matched_row['First Name'], matched_row['Last Name'], matched_row['Email'], max_similarity])

# Create a DataFrame from the results
result_df = pd.DataFrame(results, columns=['First Name (DB1)', 'Last Name (DB1)', 'Email (DB1)', 'First Name (DB2)', 'Last Name (DB2)', 'Email (DB2)', 'Similarity Score'])

# Analyze the results
threshold = 0.8  # Similarity score threshold to consider records as duplicates
duplicates_df = result_df[result_df['Similarity Score'] >= threshold]

# Print the identified duplicates
duplicates_df
```

Below is another example of querying same list of attributes from 2 different datsets and running comparison metrics
<br><br>

```python
import requests
import pandas as pd
from fuzzywuzzy import fuzz

# Define API keys
OMDB_API_KEY = "YOUR_OMDB_API_KEY"
TMDB_API_KEY = "YOUR_TMDB_API_KEY"


def fetch_data_from_omdb(movie_titles):
    """
    Fetch movie details from OMDb API.

    Args:
    movie_titles (list): List of movie titles.

    Returns:
    pd.DataFrame: Dataframe containing details from OMDb API.
    """
    omdb_data = []
    for title in movie_titles:
        response = requests.get(f"http://www.omdbapi.com/?t={title}&apikey={OMDB_API_KEY}")
        data = response.json()
        omdb_data.append({
            "Title": data.get("Title"),
            "Director": data.get("Director"),
            "Actors": data.get("Actors")
        })
    return pd.DataFrame(omdb_data)


def fetch_data_from_tmdb(movie_titles):
    """
    Fetch movie details from TMDb API.

    Args:
    movie_titles (list): List of movie titles.

    Returns:
    pd.DataFrame: Dataframe containing details from TMDb API.
    """
    tmdb_data = []
    for title in movie_titles:
        response = requests.get(f"https://api.themoviedb.org/3/search/movie?query={title}&api_key={TMDB_API_KEY}")
        data = response.json()
        if data['results']:
            movie_id = data['results'][0]['id']
            movie_details_response = requests.get(f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={TMDB_API_KEY}&append_to_response=credits")
            movie_details = movie_details_response.json()
            director = ', '.join([crew['name'] for crew in movie_details['credits']['crew'] if crew['job'] == 'Director'])
            actors = ', '.join([cast['name'] for cast in movie_details['credits']['cast'][:5]])
            tmdb_data.append({
                "Title": movie_details.get("original_title"),
                "Director": director,
                "Actors": actors
            })
    return pd.DataFrame(tmdb_data)


def compare_data(omdb_data, tmdb_data):
    """
    Compare data from OMDb and TMDb APIs and calculate the Levenshtein distance.

    Args:
    omdb_data (pd.DataFrame): Dataframe with OMDb data.
    tmdb_data (pd.DataFrame): Dataframe with TMDb data.

    Returns:
    pd.DataFrame: Dataframe with comparison results.
    """
    comparison_data = []
    for omdb_row, tmdb_row in zip(omdb_data.iterrows(), tmdb_data.iterrows()):
        comparison_data.append({
            "Title": omdb_row[1]['Title'],
            "OMDb_Director": omdb_row[1]['Director'],
            "TMDb_Director": tmdb_row[1]['Director'],
            "Director_Distance": fuzz.token_sort_ratio(omdb_row[1]['Director'], tmdb_row[1]['Director']),
            "OMDb_Actors": omdb_row[1]['Actors'],
            "TMDb_Actors": tmdb_row[1]['Actors'],
            "Actors_Distance": fuzz.token_sort_ratio(omdb_row[1]['Actors'], tmdb_row[1]['Actors'])
        })
    return pd.DataFrame(comparison_data)


# List of movie titles
movie_titles = ["Inception", "The Dark Knight", "Memento"]

# Fetch data from APIs
omdb_data = fetch_data_from_omdb(movie_titles)
tmdb_data = fetch_data_from_tmdb(movie_titles)

# Compare data and get comparison dataframe
comparison_data = compare_data(omdb_data, tmdb_data)

# Print comparison data
comparison_data
```

<br><br>
As you noticed in the above example we used `fuzz.token_sort_ratio` which leads us to the below discussion of the difference between the 3 text comparison methods we have seen so far.

1. <b>Levenshtein Distance</b> (`lev.distance`):
- <b>Definition:</b> It is the minimum number of single-character edits (insertions, deletions, or substitutions) required to change one word into the other.
- <b>Usage:</b> It is often used when the comparison between strings needs to be exact, considering the position of each character. It's suitable when the strings are short and of similar lengths.
- <b>Drawback:</b> It can give a high distance value for strings with small differences but in different orders.

2. <b>FuzzyWuzzy Ratio</b> (`fuzz.ratio`):
- <b>Definition:</b> It uses Levenshtein Distance to calculate the similarity between two strings. The output is a ratio of the similarity between the two strings.
- <b>Usage:</b> Similar to lev.distance, it's used for comparing strings with small lengths and where character positions are vital.
- <b>Drawback:</b> It also struggles with strings having elements in different orders.

3. <b>FuzzyWuzzy Token Sort Ratio</b> (fuzz.token_sort_ratio):
- <b>Definition:</b> It tokenizes the strings (splits the string into words), sorts the tokens alphabetically, and then joins them back into a string. After this, it uses the fuzz.ratio to calculate the similarity.
- <b>Usage:</b> It is more robust when the strings have the same words but in different orders. It's particularly useful in cases like comparing sentences or paragraphs where the word order might not be exact.
- <b>Advantage:</b> It can handle strings with elements in different orders more efficiently compared to the other two methods mentioned.


# Other text comparison algorithms

1. <b>Jaccard Similarity:</b>
- <b>Definition:</b> This method calculates the similarity between two sets. It is defined as the size of the intersection divided by the size of the union of the sets.
- <b>Usage:</b> Often used for comparing the similarity between two documents or strings represented as sets of tokens.

2. <b>Cosine Similarity:</b>
- <b>Definition:</b> This method measures the cosine of the angle between two vectors in a multidimensional space. It's often used in text analysis to measure the similarity between two documents represented as vectorized TF-IDF scores.
- <b>Usage:</b> Generally used in text mining for comparing document similarity.

3. <b>Hamming Distance:</b>
- <b>Definition:</b> It measures the minimum number of substitutions required to change one string into the other, or the minimum number of errors that could have transformed one string into the other.
- <b>Usage:</b> It is used mainly for error detection or error correction. It is applied to strings of equal length.

4. <b>Jaro-Winkler Distance:</b>
- <b>Definition:</b> It is a variant of the Jaro distance metric which is a type of string edit distance. The Jaro-Winkler distance gives higher scores for strings that match from the beginning.
- <b>Usage:</b> It is widely used in record linkage applications to detect duplicate records.

5. <b>Sørensen–Dice coefficient:</b>
- <b>Definition:</b> This is another method to find the similarity between two sets, calculated as twice the size of the intersection of the sets divided by the sum of the sizes of the two sets.
- <b>Usage:</b> Similar to Jaccard similarity, it is used to compare the similarity between two sets of tokens.

6. <b>Longest Common Subsequence (LCS):</b>
- <b>Definition:</b> It finds the longest subsequence common to all sequences in a set of sequences (often just two sequences).
- <b>Usage:</b> It can be used to find the similarity between two strings by finding the longest common subsequence and considering the length of LCS and the length of the strings.

7. <b>Sequence Matchers (like Ratcliff/Obershelp algorithm):</b>
- <b>Definition:</b> It is a heuristic to find the longest contiguous matching subsequence that contains no “junk” elements.
- <b>Usage:</b> Python's difflib module contains SequenceMatcher which implements this algorithm and can be used to find similarities between two strings.

Following is an exmaple for usage of each of the text comparison algorithms discussed above and comparing them. 
<br><br>

```python
import random
import pandas as pd
from fuzzywuzzy import fuzz
import distance
from matplotlib import pyplot as plt
import jellyfish

# Function to calculate Jaccard Similarity
def jaccard_similarity(str1, str2):
    """
    Calculate the Jaccard similarity between two strings.
    
    The Jaccard similarity is defined as the size of the intersection 
    divided by the size of the union of two sets.
    
    Parameters:
    str1, str2 (str): Two strings to be compared.
    
    Returns:
    float: Jaccard similarity score.
    """
    a = set(str1)
    b = set(str2)
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))

# Function to calculate Sørensen–Dice Coefficient
def sorensen_dice_coefficient(str1, str2):
    """
    Calculate the Sørensen–Dice coefficient between two strings.
    
    The Sørensen-Dice coefficient is calculated as 2 times the size of 
    the intersection of the sets divided by the sum of the sizes of 
    the two sets.
    
    Parameters:
    str1, str2 (str): Two strings to be compared.
    
    Returns:
    float: Sørensen–Dice coefficient score.
    """
    a = set(str1)
    b = set(str2)
    c = a.intersection(b)
    return 2. * len(c) / (len(a) + len(b))

# Function to calculate Longest Common Subsequence
def longest_common_subsequence(str1, str2):
    """
    Calculate the length of the Longest Common Subsequence (LCS) 
    between two strings.
    
    The LCS is the longest sequence of characters that appear left-to-right
    in both strings (but not necessarily in a contiguous block).
    
    Parameters:
    str1, str2 (str): Two strings to be compared.
    
    Returns:
    int: Length of the LCS.
    """
    m = len(str1)
    n = len(str2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(m + 1):
        for j in range(n + 1):
            if i == 0 or j == 0:
                dp[i][j] = 0
            elif str1[i - 1] == str2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]

# Step 1: Generate a dataset with pairs of strings with varying levels of similarity
names = [
    "John Doe", "Jane Smith", "Robert Johnson", "Michael Brown", "Emily Davis",
    "Sarah Wilson", "David Taylor", "James Anderson", "Mary Thomas", "Patricia White"
]

emails = [
    "john.doe@example.com", "jane.smith@example.com", "robert.johnson@example.com", 
    "michael.brown@example.com", "emily.davis@example.com", "sarah.wilson@example.com",
    "david.taylor@example.com", "james.anderson@example.com", "mary.thomas@example.com", 
    "patricia.white@example.com"
]

# Create variations of the base texts to mimic real-world data
dataset = []
for name, email in zip(names, emails):
    altered_name = name.replace(random.choice(name.split()), random.choice(names).split()[0])
    altered_email = email.replace(email.split("@")[0], random.choice(emails).split("@")[0])
    dataset.append((name, altered_name, email, altered_email))

# Convert dataset to a pandas dataframe
df = pd.DataFrame(dataset, columns=['Original Name', 'Altered Name', 'Original Email', 'Altered Email'])

# Step 2: Perform text comparison and add results to dataframe
for index, row in df.iterrows():
    original_name, altered_name, original_email, altered_email = row['Original Name'], row['Altered Name'], row['Original Email'], row['Altered Email']
    
    if isinstance(original_name, str) and isinstance(altered_name, str):
        df.at[index, 'Name Fuzz Ratio'] = fuzz.ratio(original_name, altered_name)
        df.at[index, 'Name Token Sort Ratio'] = fuzz.token_sort_ratio(original_name, altered_name)
        df.at[index, 'Name Lev Distance'] = distance.levenshtein(original_name, altered_name)
        df.at[index, 'Name Jaccard Similarity'] = jaccard_similarity(original_name, altered_name)
        df.at[index, 'Name Jaro-Winkler Distance'] = jellyfish.jaro_winkler(original_name, altered_name)
        df.at[index, 'Name Sørensen-Dice Coefficient'] = sorensen_dice_coefficient(original_name, altered_name)
        df.at[index, 'Name LCS'] = longest_common_subsequence(original_name, altered_name)
    
    if isinstance(original_email, str) and isinstance(altered_email, str):
        df.at[index, 'Email Fuzz Ratio'] = fuzz.ratio(original_email, altered_email)
        df.at[index, 'Email Token Sort Ratio'] = fuzz.token_sort_ratio(original_email, altered_email)
        df.at[index, 'Email Lev Distance'] = distance.levenshtein(original_email, altered_email)
        df.at[index, 'Email Jaccard Similarity'] = jaccard_similarity(original_email, altered_email)
        df.at[index, 'Email Jaro-Winkler Distance'] = jellyfish.jaro_winkler(original_email, altered_email)
        df.at[index, 'Email Sørensen-Dice Coefficient'] = sorensen_dice_coefficient(original_email, altered_email)
        df.at[index, 'Email LCS'] = longest_common_subsequence(original_email, altered_email)

# Step 4: Visualize the comparison results using scatter plots
labels = ['Fuzz Ratio', 'Token Sort Ratio', 'Levenshtein Distance', 'Jaccard Similarity', 'Jaro-Winkler Distance', 'Sørensen–Dice Coefficient', 'Longest Common Subsequence']
name_results = [label for label in labels if 'Name ' + label in df.columns]
email_results = [label for label in labels if 'Email ' + label in df.columns]

for label in name_results:
    plt.figure()
    plt.scatter(range(len(df)), df['Name ' + label])
    plt.title('Name ' + label)
    plt.xlabel('Data Pair Index')
    plt.ylabel('Score')
    plt.grid(True)
    plt.show()

for label in email_results:
    plt.figure()
    plt.scatter(range(len(df)), df['Email ' + label])
    plt.title('Email ' + label)
    plt.xlabel('Data Pair Index')
    plt.ylabel('Score')
    plt.grid(True)
    plt.show()
```

<b>Note:</b> `Hamming Distance` was not included in the above example as its a special case.

The Hamming distance is a measure of the difference between two strings of equal length, calculated as the number of positions at which the corresponding elements are different. Given its requirement for strings of equal length, it is not as flexible as some other distance metrics when it comes to comparing arbitrary strings.

It can be used for:
- <b>Error Detection and Correction:</b> In coding theory, Hamming distance is used to find errors in codes and correct them. It's mainly used in computer communication protocols.

- <b>DNA Sequence Comparison:</b> In bioinformatics, Hamming distance can be used to quantify the genetic distance between two DNA sequences of equal length.

- <b>Image Processing:</b> It can be used in image processing to compare two images (or binary strings representing images) of the same size to find the differences between them.



```python
"""Example to showcase DNA Sequence comparison using Hamming distance"""

import random
import pandas as pd
from matplotlib import pyplot as plt

# Step 1: Generate an extensive dataset of DNA sequences of equal length
dataset = []
for _ in range(1000):
    str1 = ''.join(random.choice('ACGT') for _ in range(100))  # 100-base DNA sequences
    str2 = ''.join(random.choice('ACGT') for _ in range(100))
    dataset.append((str1, str2))

# Convert dataset to a pandas dataframe
df = pd.DataFrame(dataset, columns=['Sequence 1', 'Sequence 2'])

# Step 2: Define a function to calculate Hamming distance
def hamming_distance(str1, str2):
    return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))

# Step 3: Calculate Hamming distance for each pair of strings in the dataset
hamming_distance_results = []
for index, row in df.iterrows():
    str1, str2 = row['Sequence 1'], row['Sequence 2']
    hamming_distance_results.append(hamming_distance(str1, str2))

# Add Hamming distance results to the dataframe
df['Hamming Distance'] = hamming_distance_results

# Step 4: Visualize the Hamming distance results using a histogram
plt.hist(hamming_distance_results, bins=20, edgecolor='black')
plt.title('Hamming Distance')
plt.xlabel('Hamming Distance')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
